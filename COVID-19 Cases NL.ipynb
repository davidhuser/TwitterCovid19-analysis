```python
import pandas as pd

# https://www.rivm.nl/nieuws/actuele-informatie-over-coronavirus

data = [
{'date': '2020-02-27', 'confirmed': 1},
{'date': '2020-02-28', 'confirmed': 2},
{'date': '2020-02-29', 'confirmed': 6},
{'date': '2020-03-01', 'confirmed': 9},
{'date': '2020-03-02', 'confirmed': 19},
{'date': '2020-03-03', 'confirmed': 24},
{'date': '2020-03-04', 'confirmed': 38},
{'date': '2020-03-05', 'confirmed': 82},
{'date': '2020-03-06', 'confirmed': 128},
{'date': '2020-03-07', 'confirmed': 188},
{'date': '2020-03-08', 'confirmed': 264},
{'date': '2020-03-09', 'confirmed': 321},
{'date': '2020-03-10', 'confirmed': 382},
{'date': '2020-03-11', 'confirmed': 503},
{'date': '2020-03-12', 'confirmed': 614},
{'date': '2020-03-13', 'confirmed': 804}
]

df = pd.DataFrame(data)
df['date'] = pd.to_datetime(df['date'])
df['date'] = df.date.dt.date
df['growth'] = df['confirmed'].pct_change()
df.info
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import rcParams

plt.title("Confirmed cases of COVID-19 in Netherlands (as of March 13 2020)")
sns.set(style="dark")
palette = sns.color_palette("bright", 3)

g = sns.barplot(data=df, x='date', y='confirmed', color='blue')
g.set(ylabel=None)
plt.gcf().autofmt_xdate()

plt.savefig('confirmed_cases_nl.png')
```
